```python
import random as rm
```

```python
def gamewin(comp,you):
    if comp==you:
        return
    elif comp=="s":
        if you=="w":
            return False
        elif comp=="g":
            return True
        
    elif comp=="w":
        if you=="s":
            return False
        elif you=="g":
            return True
        
    elif comp=="g":
        if you=="s":
            return False
        elif you=="w":
            return True


print("Its computer's turn:\nSnake(s) Water(w) or Gun(g)")
print("*****************************************************************")
randno=rm.randint(1,3)
if randno==1:
    comp="s"
elif randno==2:
    comp="w"
elif randno==3:
    comp="g"
you=input("Now its your turn:\n[Note kindly choose without space]\nSnake(s) Water(w) or Gun(g): ")
a=gamewin(comp,you)


if a==None:
    print("Its a tie!!! ")
elif a:
    print("You win!!!! ")
else:
    print("You lose!!! ")

print("*****************************************************************")    
print(f"Your Choice is: {you}")
print(f"Choice of computer is: {comp}")
```
