### Defining X and Y varaibles 

```python
x=df.drop(columns="Outcome",axis=1)
y=df["Outcome"]
```

```python
x.head()
```

```python
y.tail()
```

### Train and Test Spliting 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

#### Fitting model

```python
log_rig=LogisticRegression()
log_rig.fit(x_train,y_train)
```

#### Predicting values using .predict function

```python
log_pred=log_rig.predict(x_test)
```

#### Accuracy score for measuring accuracy


```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,log_pred)
```

#### Confusion matrics for True Positive, False Positive, False Negative and True Positive
#### Classification report for Precision, Recall, F1-Score

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,log_pred))
print(classification_report(y_test,log_pred))
```

## Support Vector Machine (SVM)

```python
from sklearn import svm
```

```python
svm_model=svm.SVC(kernel="linear")
```

#### Fitting model and predicting with model

```python
svm_model.fit(x_train,y_train)
```

#### Predicting values using .predict function
#### Confusion matrics for True Positive, False Positive, False Negative and True Positive
#### Classification report for Precision, Recall, F1-Score

```python
svm_pred=svm_model.predict(x_test)
```

```python
accuracy_score(y_test,svm_pred)
```

```python
print(confusion_matrix(y_test,svm_pred))
print(classification_report(y_test,svm_pred))
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

#### Fitting model and predicting with model

```python
rfc_model=RandomForestClassifier()
rfc_model.fit(x_train,y_train)
```

#### Predicting values using .predict function
#### Confusion matrics for True Positive, False Positive, False Negative and True Positive
#### Classification report for Precision, Recall, F1-Score

```python
rfc_pred=rfc_model.predict(x_test)
```

```python
print("Accuracy score is: ",accuracy_score(y_test,rfc_pred))
```

```python
print(confusion_matrix(y_test,rfc_pred))
print(classification_report(y_test,rfc_pred))
```
